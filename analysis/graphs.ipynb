<h1>Assignment 3 Graphs</h1>

```python
import seaborn as sns
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def annotate(data, **kws):
    #r, p = sp.stats.pearsonr(data['ID'], data['Time (ms)'])
    
    slope, intercept, r, p_value, std_err = stats.linregress(data['ID'],data['Time (ms)'])

    ax = plt.gca()
    ax.text(.05, .8, 'r2={:.2f}'.format(r*r),  #'y = ' + str(round(intercept,3)) + ' + ' + str(round(slope,3)) + 'x',
            transform=ax.transAxes)
```

```python
all_data = pd.read_csv("./results.csv")
```

```python
all_data.columns
```

<h2>No Distortion</h2>

```python
no_distortion = all_data[all_data['Condition Name'] == 'No distortion - control']
no_distortion = no_distortion[no_distortion['Trial']>5]
```

```python
a = sns.lmplot(data=no_distortion, x="ID", y="Time (ms)")    
a.map_dataframe(annotate)
a.savefig("no_distortion.png")
```

<h2>Low Distortion : 0.5</h2>

```python
low_distortion = all_data[all_data['Condition Name'] == 'Minimal distortion']
low_distortion = low_distortion[low_distortion['Trial']>5]
b = sns.lmplot(data=low_distortion, x="ID", y="Time (ms)")    
b.map_dataframe(annotate)
b.savefig("low_distortion.png")
```
